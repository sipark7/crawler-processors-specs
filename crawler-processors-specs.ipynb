```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import urllib3
import certifi

from __future__ import print_function
```

```python
# fixes certification error
http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED',
                           ca_certs=certifi.where())

# sample processor information
url_t1 = "https://ark.intel.com/products/137980/Intel-Core-i5-8269U-Processor-6M-Cache-up-to-4-20-GHz-"
url_t2 = "https://ark.intel.com/products/130409/Intel-Core-i7-8809G-Processor-with-Radeon-RX-Vega-M-GH-graphics-8M-Cache-up-to-4-20-GHz-"

def get_soup(url):
    response = http.request('Get', url)
    soup = BeautifulSoup(response.data, "lxml")
    return soup
```

```python
# retreive the main section where data is scraped
def get_tables(url):
    test_soup = get_soup(url)
    return test_soup.find_all('div', attrs={'id':'tab-blade-1-0'}) # specifications section
```

```python
# test_tables = test_soup.find_all('div', attrs={'class':'blade-inside'})
# previously used to get all sections with data, but included some that were outside specifications
```

```python
# dataframe function - retrieve labels and values of a subsection
def get_df(i, tables):
    table_labels = tables[i].find_all('span', attrs={'class':'label'})
    table_values = tables[i].find_all('span', attrs={'class':'value'})
    x = [label.get_text().strip() for label in table_labels]
#     print(x)
    y = [value.get_text().strip() for value in table_values]
#     print(y)
    df_xy = pd.DataFrame([dict(zip(x,y))])
    return df_xy
```

```python
for i in range(len(href_list)):
    soup = get_soup('https://ark.intel.com' + href_list[i])
    table_soup = soup.find('tbody')
    proc_table = table_soup.find_all('a', href=True)
    for j in proc_table:
        href_products_list.append(j['href'])
```

```python
len(href_products_list)
```

```python
count = 0
for i in range(len(href_products_list)):
    table = get_tables("https://ark.intel.com" + href_products_list[i])
#     print(len(table))
    for j in range(len(table)):
        specs = get_df(j, table)
        print(count)
        if count == 0:
            total = specs
        else:
            total = total.append(specs, ignore_index=True, sort=True)
        count = count + 1
```
